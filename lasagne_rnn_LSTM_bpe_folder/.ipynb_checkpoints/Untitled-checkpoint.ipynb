```python
words=[]
word=""
for line in list(open("/home/nbuser/library/Warhammer.txt",encoding='cp1251')):
    for i in line:
        if(i==' ' or i=='\n'):
            words.append(word)
            word=''
        else:
            word+=i
f = open('Words_warhammer.txt', 'w',encoding='cp1251')
for i in words:
    f.write(i)
    f.write("\n")
f.close()
```
